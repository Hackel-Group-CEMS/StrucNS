```python
import pandas as pd
import os

# --- Configuration ---
BASE_INPUT_DIR = "/scratch.global/hackelb/mulli468/Tsuboyama_analysis/processing_data/feature_datasets/StrucNS_sets/Training/base_model/case1/predictions"

# These are the original source files
INPUT_CSVS = [
    "Test_Set_1_predictions.csv",
    "Test_Set_2_predictions.csv",
    "Test_Set_3_predictions.csv"
]

def update_output_with_ddg(input_csv_name):
    input_path = os.path.join(BASE_INPUT_DIR, input_csv_name)
    
    # Matching your specific output naming: scored_esm1v_Test_Set_1_predictions.csv
    output_path = f"scored_esmIF_{input_csv_name}"

    if not os.path.exists(output_path):
        print(f"Skipping: {output_path} does not exist in current directory.")
        return

    print(f"Updating {output_path} with data from {input_csv_name}...")

    # 1. Load the data
    df_input = pd.read_csv(input_path)
    df_output = pd.read_csv(output_path)

    # 2. Create Mappings using 'file' as the unique key
    # Using .get() to handle cases where columns might be missing or differently named
    ddg_map = df_input.set_index('file')['ddG'].to_dict()
    dg_mut_map = df_input.set_index('file')['dg_mut'].to_dict()
    dg_wt_map = df_input.set_index('file')['dg_wt'].to_dict()

    # 3. Fill the columns in the output dataframe
    # map() looks up the 'file' ID in our dictionaries and places the correct value
    df_output['ddG'] = df_output['file'].map(ddg_map)
    df_output['dg_mut'] = df_output['file'].map(dg_mut_map)
    df_output['dg_WT'] = df_output['file'].map(dg_wt_map)

    # 4. Save the updated CSV (overwriting the output)
    df_output.to_csv(output_path, index=False)
    print(f"Successfully updated {output_path} (Rows: {len(df_output)})")

if __name__ == "__main__":
    for csv_name in INPUT_CSVS:
        try:
            update_output_with_ddg(csv_name)
        except Exception as e:
            print(f"Failed to update {csv_name}: {e}")
```
